```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a pandas DataFrame
df = pd.read_csv('c://users/alliata/survey1_3.csv')

# Function to create a bar plot for a given question
def plot_question(question):
    yes_count = df[question].str.contains('Yes').sum()
    no_count = df[question].str.contains('No').sum()
    
    plt.figure(figsize=(8, 6))
    plt.bar(['Yes', 'No'], [yes_count, no_count])
    plt.title(question)
    plt.xlabel('Answer')
    plt.ylabel('Count')
    plt.show()

# Plot each question
plot_question('Is your company using AI right now in their business processes? ')
plot_question('Do you think AI will replace your job?  ')
plot_question('Do you think AI will significantly change your job? ')
plot_question('Do you think AI is mostly hype? ')
plot_question('Do you think AI will have a real impact on industry?')
plot_question('Do you think AI will have a real impact on society?')
plot_question('Do you use or plan to use AI to generate user stories? ')
plot_question('Do you use or plan to use AI to either explain or summarize user stories?')
plot_question('Do you use or plan to use AI to help you run your retrospectives?')
plot_question('Do you use or plan to use AI to create status reports?')
plot_question('Do you use or plan to use AI to summarize meetings and/or create action items?   ')
plot_question('Do you use or plan to use AI in other parts of your job? ')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a pandas DataFrame
df = pd.read_csv('c://users/alliata/survey1_3.csv')

# Group the data by Gender and count the Yes/No answers for each question
gender_counts = df.groupby(['Gender', 'Is your company using AI right now in their business processes? '])['Title'].count().unstack(fill_value=0)
gender_counts = gender_counts.reindex(['Male', 'Female', 'NA'])

# Create a stacked bar plot
gender_counts.plot(kind='bar', stacked=True, figsize=(12,6))

# Set plot title and axis labels
plt.title('AI Usage and Perception by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='AI Usage', loc='upper left')

# Show the plot
plt.show()

```
